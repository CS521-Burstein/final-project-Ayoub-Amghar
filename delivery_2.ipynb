```python
from pandasql import sqldf
sql_pd_dataframe = lambda q: sqldf(q, globals())
```

```python
# show the saleprice where the row is not null
q = """SELECT SalePrice 
       FROM housing_dataframe WHERE SalePrice IS NOT NULL;"""

names = sql_pd_dataframe(q)
names
```

```python
#Show the houses that are built after 2000 
q = """SELECT YearBuilt
       FROM housing_dataframe WHERE YearBuilt > 2000 ORDER BY YearBuilt;"""

house_after_2000 = sql_pd_dataframe(q)
house_after_2000 
```
