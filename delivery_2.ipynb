# Histograms using seaborn

```python
import seaborn as sns
from scipy.stats import norm

sns.distplot(housing_dataframe['SalePrice'], fit = norm, color = "purple", bins = 50)
plt.title('Housing Sale Price', fontsize=15)
```

```python
plt.figure(figsize=(15,6))
sns.histplot(x= "SalePrice",
             data=housing_dataframe,
              bins=30,
              hue="GarageCars",
              element="step",
             color=['red','purple','pink','blue','yellow','green'],
              kde=True)
plt.savefig("Histograms_saleprice_garageCar.jpg")
```
